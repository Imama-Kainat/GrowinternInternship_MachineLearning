```python
# Description: This program attempts to perdict the future price of stock
```

```python
#import the libraries
import pandas as pd
import numpy as np
from sklearn.ensemble import RandomForestRegressor
import matplotlib.pyplot as plt
```

```python
#collect and clean data: df ->data frame
df = pd.read_csv('stockdata.csv')
df = df.dropna()
#look at the data
df
```

```python
# show the data visually
df.plot(x="Date", y="Close")
plt.xticks(rotation=45)
```

```python
#Create the model
model= RandomForestRegressor()
```

```python
#train the model
X=df[['Open', 'High','Low','Volume']]
X= X[:int(len(df)-1)]
y=df['Close']
y=y[:int(len(df)-1)]
model.fit(X,y) # training the model
```

```python
#Test the model
predicions=model.predict(X)
print('The model score is :', model.score(X,y))
```

```python
#Make the predictions
new_data=df[['Open', 'High','Low','Volume']].tail(1)
prediction = model.predict(new_data)
print('The model predicts the last row or day to be:',prediction)
print('Actual value is:',df[['Close']].tail(1).values[0][0])
```
